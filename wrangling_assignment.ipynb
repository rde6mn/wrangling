  1. Read the abstract. What is this paper about?
  
  The paper is about using the tidy data framework to easily turn un-tidy datsets into tidy ones.
  
  2. Read the introduction. What is the "tidy data standard" intended to accomplish?

  The tidy data standard intends to provide a standard way to organize data values within a dataset.

  3. Read the intro to section 2. What does this sentence mean: "Like families, tidy datasets are all alike but every messy dataset is messy in its own way." What does this sentence mean: "For a given dataset, it’s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general."

  The first sentence means that no two datasets have the exact same problems within the data/things that make it messy.

  The second sentence means that it's easy to figure out what the observations (each individual record) and variables (columns) are, but the hard part is figuring our what they mean.

  4. Read Section 2.2. How does Wickham define values, variables, and observations?

  Wickham says that values make up datasets. Variables are values that all measure the same underlying attribute. Observations are all values measured on the same unit.
  
  5. How is "Tidy Data" defined in section 2.3?

  Tidy data is defined as data where variables form columns, observations form rows, and types of observational units form tables.

  6. Read the intro to Section 3 and Section 3.1. What are the 5 most common problems with messy datasets? Why are the data in Table 4 messy? What is "melting" a dataset?

  The five most common problems are: column headers are values instead of variable names, there are multiple variables in a column, varibles are stored in rows and in columns, there are many types of observational units in one table, and a single observational unit is in multiple tables.

  The data in table 4 is messy because the variables are stored in rows and columns.

  Melting a dataset means stacking it.

  7. Why, specifically, is table 11 messy but table 12 tidy and "molten"?

  8. Read Section 6. What is the "chicken-and-egg" problem with focusing on tidy data? What does Wickham hope happens in the future with further work on the subject of data wrangling?

  The chicken and the egg problem is that tidy data is only as useful as the tools that are used to work with it. Due to this, it can be the case where the workflow is not improved by changing data structures or tools.

  Wickham hopes that others will build upon his framework to develop better data storage strategies and tools.

```python
import pandas as pd
import numpy as np
```

```python
bnb = pd.read_csv('/content/wrangling/assignment/data/airbnb_hw.csv')
bnb.sort_values(by=['Price'], ascending=False)
```

For price formatting of 999, there are commas. You can't have commas in numeric variables, so to clean this, I will replace commas with nothing to remove them, and then, I will convert the column to be a numeric type.

```python
bnb['Price'] = bnb['Price'].str.replace(',','') #removed commas to prepare it to be converted to a numeric type
bnb['Price'] = pd.to_numeric(bnb['Price'], errors = 'coerce') #converts column to numeric type
bnb.sort_values(by=['Price'], ascending=False)
```

```python
print(sum(bnb['Price'].isnull())) #prints number of missing values
```

As shown above, there are 0 missing values

```python
shark = pd.read_csv('/content/wrangling/assignment/data/sharks.csv')
print(shark['Type'].unique())
```

```python
#THese don't add value to the dataset, and it is unclear what they mean
shark['Type'] = shark['Type'].replace(['Questionable','Unconfirmed','Unverified','Invalid'], np.nan)

#THese are all boat/sea related disasters, so I've combined them into one column
shark['Type'] = shark['Type'].replace(['Watercraft', 'Boating', 'Sea Disaster','Boat', 'Boatomg'], 'Boats')

print(shark['Type'].unique())
```

```python
#the float values are ugly, so i will turn them into ints
trial['ImposedSentenceAllChargeInContactEvent'] = pd.to_numeric(trial['ImposedSentenceAllChargeInContactEvent'], errors='coerce')
print(np.sum(trial['ImposedSentenceAllChargeInContactEvent'].isnull()))
```

```python
print(pd.crosstab(trial['ImposedSentenceAllChargeInContactEvent'].isnull(), trial['SentenceTypeAllChargesAtConvictionInContactEvent']))

#replaces 'ImposedSentenceAllChargeInContactEvent' values based on 'SentenceTypeAllChargesAtConvictionInContactEvent' values
trial['ImposedSentenceAllChargeInContactEvent'] = trial['ImposedSentenceAllChargeInContactEvent'].mask(trial['SentenceTypeAllChargesAtConvictionInContactEvent']==4,0)
trial['ImposedSentenceAllChargeInContactEvent'] = trial['ImposedSentenceAllChargeInContactEvent'].mask(trial['SentenceTypeAllChargesAtConvictionInContactEvent']==9,np.nan)
```
